```python
# ==========================================
# Task 1: Data Immersion & Wrangling
# Dataset: Ecommerce Sales Data
# ==========================================

# Step 1: Import libraries
import pandas as pd
import numpy as np


# Step 2: Load dataset
df = pd.read_csv("data.csv", encoding="latin1")

print("First 5 rows:")
print(df.head())


# Step 3: Dataset overview
print("\nDataset shape:", df.shape)

print("\nColumn names:")
print(df.columns)

print("\nDataset information:")
print(df.info())

print("\nStatistical summary:")
print(df.describe())


# Step 4: Data Dictionary
data_dictionary = pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes.values,
    "Meaning": [
        "Unique invoice number",
        "Product code",
        "Product description",
        "Number of items purchased",
        "Date and time of purchase",
        "Price per product",
        "Unique customer ID",
        "Customer country"
    ]
})

print("\nData Dictionary:")
print(data_dictionary)


# Step 5: Check missing values
print("\nMissing values in each column:")
print(df.isnull().sum())


# Remove missing CustomerID rows
df = df.dropna(subset=["CustomerID"])


# Step 6: Check duplicate rows
print("\nDuplicate rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()


# Step 7: Convert InvoiceDate column to datetime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])


# Step 8: Remove invalid values

# Remove negative quantity (product returns)
df = df[df["Quantity"] > 0]

# Remove zero or negative price
df = df[df["UnitPrice"] > 0]


# Step 9: Feature Engineering

# create total sales column
df["TotalSales"] = df["Quantity"] * df["UnitPrice"]

# extract year
df["Year"] = df["InvoiceDate"].dt.year

# extract month
df["Month"] = df["InvoiceDate"].dt.month

# extract day
df["Day"] = df["InvoiceDate"].dt.day


# Step 10: Final dataset check
print("\nFinal dataset preview:")
print(df.head())

print("\nFinal dataset shape:", df.shape)


# Step 11: Save cleaned dataset
df.to_csv("cleaned_data.csv", index=False)

print("\nCleaned dataset saved successfully as cleaned_data.csv")
```
